## Calculating Distance(Kilometers) and Speed (km/hr)

```python
#%pip install geopy
#Importing geopy library
from geopy.distance import geodesic

#Reseting index
Final_df = Final_df.reset_index(drop=True)


#Creating List and appending "0" distance for the first tuple of first taxi
Final_distance = []
Final_distance.append(0)

#Running For loop for each tuple of taxi's
for i in range(0,158225):
  
  #Getting the Taxi Id for the current row and next row
  current_taxi = Final_df.Taxi_id[i]
  next_taxi = Final_df.Taxi_id[i+1]
  

  #Checking the condition for same taxi id - if it is same then only it will calculate distance
  if (current_taxi == next_taxi): 
     
     #Getting Latitude and Longitude for current taxi
     loc_1 = (Final_df.latitude[i], Final_df.longitude[i]) 

     #Getting Latitude and Longitude for next taxi
     loc_2 = (Final_df.latitude[i+1], Final_df.longitude[i+1])
    
     #Append the geodesic distance in kilometers calculated for 2 different locations for the same taxi. 
     Final_distance.append(geodesic(loc_1, loc_2).km)
  
  #If Taxi id for current and next taxi are not same then "0" will be appended to list to represent new Taxi
  else:
     Final_distance.append(0)

```

```python
#Adding list to Dataframe column
Final_df['Distance'] = Final_distance
```
